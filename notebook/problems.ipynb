### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import pandas as pd

cantExp=1000  #count experiments
exito=0       #keep track the number of times that the sum >7 or even number

#Funtion to roll the dice.
for i in range(1,cantExp):
    sum = np.random.randint(1,7) +  np.random.randint(1,7)
    if sum>7 or sum%2==0:
        exito+=1

prob=exito/cantExp
print("Probabilidad: " + str(prob))
```

```python
#OTHER OPTION TO CALCULATE

import numpy as np 
dice_space =np.arange(1,7)# Arange will take the interval betweeen (a, b-1)
dice_space

cantExp=10000  #count experiments

#simulate one dice
np.random.seed(27) # Make your findings reproducible
trial_1 = np.random.choice(dice_space, cantExp)
#simulate second dice
np.random.seed(30) # Make your findings reproducible
trial_2 = np.random.choice(dice_space, cantExp)
#sum
trial = trial_1 + trial_2
#filter
res = trial[(trial % 2 == 0) | (trial >7)]
display(res)

prob=len(res)/cantExp
print("Probabilidad: " + str(prob))
```
