### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import pandas as pd

cantExp=1000  #count experiments
exito=0       #keep track the number of times that the sum >7 or even number

#Funtion to roll the dice.
for i in range(1,cantExp):
    sum = np.random.randint(1,7) +  np.random.randint(1,7)
    if sum>7 or sum%2==0:
        exito+=1

prob=exito/cantExp
print("Probabilidad: " + str(prob))
```

```python
#OTHER OPTION TO CALCULATE

import numpy as np 
dice_space =np.arange(1,7)# Arange will take the interval betweeen (a, b-1)
dice_space

cantExp=10000  #count experiments

#simulate one dice
np.random.seed(27) # Make your findings reproducible
trial_1 = np.random.choice(dice_space, cantExp)
#simulate second dice
np.random.seed(30) # Make your findings reproducible
trial_2 = np.random.choice(dice_space, cantExp)
#sum
trial = trial_1 + trial_2
#filter
res = trial[(trial % 2 == 0) | (trial >7)]
display(res)

prob=len(res)/cantExp
print("Probabilidad: " + str(prob))
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import random
import numpy as np

#defined box contained ball
d = {}

for i in range(60):
    if i < 10:
        d[i]='W'
    elif i > 9 and i < 30:
        d[i]='R'
    else:
        d[i]='G'  

print(d)
  
```

```python
n_simulations=1000  #count experiments
cond3Wor2R=0
condAllEqual=0

for i in range(n_simulations):
    
    #make a list of the colors that we choose
    s=[]
    for i in range(5):
        s.append(d[random.randint(0,59)])  #both 0 and 59 include
    #display(s)
    #Convert it to a numpy array
    arrSelBall=np.array(s)

    #find the number of each that we picked
    ballW=np.count_nonzero(arrSelBall == 'W')
    ballR=np.count_nonzero(arrSelBall == 'R')
    ballG=np.count_nonzero(arrSelBall == 'G')
    #print(ballW,ballR,ballG)
    #Keep track if the combination met the criteria
    if (ballW==3 or ballR==2):
        cond3Wor2R+=1
    if (ballW==5 or ballR==5 or ballG==5):
        condAllEqual+=1
     
    
#print 'the probability of 3 white and 2 red is' in percentage
print('the probability of 3 white and 2 red is ', cond3Wor2R/n_simulations)
#print 'the probability of all the same color is' in percentage   
print('the probability of all the same color is',condAllEqual/n_simulations)
```
